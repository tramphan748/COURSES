##  TÃ­nh cÃ¡c lÃ¡ng giá»ng gáº§n nháº¥t cho 100 dÃ²ng Äáº§u tiÃªn cá»§a array á» má»¥c 2 sá»­ dá»¥ng má»¥c 2 vá»i Äá» Äo tÆ°Æ¡ng Äá»ng vÃ  Äá» Äo táº§n suáº¥t xuáº¥t hiá»n ngÆ°á»£c

### a. Äá» Äo tÆ°Æ¡ng Äá»ng
XÃ©t 2 báº£n ghi $\overline{X}=(x_{1}\ldots x_{d})$ vÃ   $\overline{Y}=(y_{1}\ldots y_{d})$, sá»± tÆ°Æ¡ng Äá»ng ÄÆ¡n giáº£n nháº¥t giá»¯a hai báº£n ghi nÃ y ÄÆ°á»£c xÃ¡c Äá»nh nhÆ° sau:
\begin{align}
Sim(\overline{X},\overline{Y}) = \sum_{i=1}^d S(x_i,y_i).
\end{align}
Vá»i: $S(x_i,y_i)$ lÃ  sá»± tÆ°Æ¡ng Äá»ng giá»¯a cÃ¡c giÃ¡ trá» thuá»c tÃ­nh $x_{i}, y_{i}$
 Lá»±a chá»n ÄÆ¡n giáº£n nháº¥t cho $S(x_i,y_i)$ lÃ :
\begin{cases}
S(x_i,y_i) = 1 , \, \text{if} \space x_i = y_i \\
S(x_i,y_i) = 0 , \,\text{in contrast}
\end{cases}

```python
def simple_matching_similarity(x, y):
  # Check if data records have the same length
  if np.array_equal(x, y):
    similarity = 1/len(x)
  else:
    similarity = 0
  return similarity

# Save the nearest neighbor variable of each data row
nearest_neighbors = []

# Loop through the first 100 data rows
for i in range (0,101):
  current_row = array[i]
  # Initialize the maximum similarity value and nearest neighbor
  max_simi = -1
  nearest_neighbor = None
  # Loop through other data points and calculate similarity
  for j in range(100):
    if i != j:
      similarity = simple_matching_similarity(current_row, array[j])
      if similarity > max_simi:
        max_simi = similarity
        nearest_neighbor = j
   # Save the nearest neighbor value for the current row
  nearest_neighbors.append(nearest_neighbor)

for i, neighbor in enumerate(nearest_neighbors):
  print(f"Nearest neighbor of row {i + 1} is row {neighbor + 1}")
```

## b. Äá» Äo táº§n suáº¥t xuáº¥t hiá»n ngÆ°á»£c:
Cho $p_{k}(x)$ lÃ  má»t tá» sá» cá»§a cÃ¡c báº£n ghi mÃ  thuá»c tÃ­nh thá»© k láº¥y giÃ¡ trá» x trong táº­p dá»¯ liá»u:
\begin{cases}
S(x_i,y_i) = \dfrac{1}{p_{k}(x)^2} \, , \text{if} \,\, x_i = y_i \\
S(x_i,y_i) = 0 , \text{in contrast}
\end{cases}

```python
def IFAM(x,y,k):
  # Calculate the frequency of occurrence of each value x[k] in the array x
  p_k_x = np.square(np.sum(x == x[k]))/len(x)
  # calculate the similarity
  if np.array_equal(x,y):
    similarity = 1 /(p_k_x**2)
  else:
    similarity = 0
  return similarity

# Initialize an empty list nearest_neighbors to store the nearest neighbors of each data row
nearest_neighbors = []

#  Loop through the first 100 data rows
for i in range (0,101):
  current_row = array[i]
  # Initialize the maximum similarity value and nearest neighbor
  max_simi = float('-inf')
  nearest_neighbor = None
  # Loop through other data points and calculate similarity
  for j in range(100):
    if i != j:
      for k in range(len(current_row)):
        similarity = IFAM(current_row, array[j],k)
        if similarity >= max_simi:
          max_simi = similarity
          nearest_neighbor = j
   # Save the nearest neighbor value for the current row
  nearest_neighbors.append(nearest_neighbor)

for i, neighbor in enumerate(nearest_neighbors):
  print(f"Nearest neighbor of row {i + 1} is row {neighbor + 1}")
```
